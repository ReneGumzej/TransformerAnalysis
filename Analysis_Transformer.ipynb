```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
df = pd.read_excel("/content/drive/MyDrive/Bachelorarbeit/Dataset/Vergleich_Transformer.xlsx")
df.rename(columns = {'train_runtime (sek)':'train_runtime'}, inplace = True)
```

```python
f, ax = plt.subplots(2,2, figsize=(18,12))

min_train = sns.barplot(ax=ax[0,0],x="Batch_train",y="train_avg_loss", hue="Modell", data=df)
min_runtime = sns.barplot(ax=ax[0,1],x="Batch_train",y="train_runtime", hue="Modell", data=df)
max_Bleurt_score = sns.barplot(ax=ax[1,0],x="Batch_train",y="Bleurt_score", hue="Modell", data=df)
min_duplicate = sns.barplot(ax=ax[1,1],x="Batch_train",y="Anzahl_Duplicate", hue="Modell", data=df)

```

```python
mean_train_GPT2 = np.mean([df.loc[2]["train_runtime"],df.loc[3]["train_runtime"]])
mean_train_GPT2NEO125 = np.mean([df.loc[4]["train_runtime"],df.loc[5]["train_runtime"]])
print(f"Mean runtime GPT2 over all Batches: ",mean_train_GPT2)
print(f"Mean runtime GPT2NEO125 over all Batches: ",mean_train_GPT2NEO125)
```
